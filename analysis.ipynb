```python
import coba as cb
import matplotlib.pyplot as plt
```

```python
cb.Result.from_file('./experiment1.log').plot_learners(labels=['IGW','$\epsilon$-greedy'])
```

```python
cb.Result.from_file('./experiment2.log').plot_learners(labels=['IGW','LinUCB'])
```

```python
cb.Result.from_file('./experiment3.log').plot_learners(x='n_actions',labels=['IGW','$\epsilon$-greedy'],style='-',out=None,err='se')
plt.ylim(0.1,1)
plt.xlabel("Action Count")
plt.show()
```

```python
cb.Result.from_file('./experiment3.log').plot_learners(x='n_actions',y='predict_time',labels=['IGW','$\epsilon$-greedy'],style='-',out=None)
plt.ylabel('prediction time (secs)')
plt.xlabel("Action Count")
plt.show()
```
