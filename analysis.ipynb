```python
import coba as cb
#load the data produced by the experiment
res = cb.Result.from_file('./out.log')
```

```python
#average reward by interaction index across 29 distinct datasets when the environment is stationary
res.filter_env(cycle_after=0).filter_fin().plot_learners(labels=['IGW','LinUCB','Epsilon-Greedy'])
```

```python
#difference in average reward on 29 stationary datasets (1 sample per dataset)
res.filter_env(cycle_after=0).filter_fin().plot_contrast(1,0,labels=['LinUCB Wins','IGW Wins'])
```
